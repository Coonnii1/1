```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
gme_page = requests.get(url_gme, headers=headers)

if gme_page.status_code == 200:
    print("Página descargada correctamente")
else:
    print(f"Error al descargar la página. Código de estado: {gme_page.status_code}")

gme_soup = BeautifulSoup(gme_page.text, "html.parser")

gme_tables = gme_soup.find_all("table")

df_gme_revenue = None

for table in gme_tables:
    if "GameStop Quarterly Revenue" in str(table):
        df_gme_revenue = pd.read_html(str(table))[0]
        break
if df_gme_revenue is not None:
    df_gme_revenue.columns = ["Date", "Revenue"]
    df_gme_revenue = df_gme_revenue[df_gme_revenue["Revenue"] != ""]
    # The dollar sign might have different formatting, use regex to ensure it's removed.
    df_gme_revenue["Revenue"] = df_gme_revenue["Revenue"].str.replace("[\$,]", "", regex=True).astype(float)

    print(df_gme_revenue.tail())
else:
    print("No se encontró la tabla de ingresos trimestrales de GameStop.")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")

tesla_data.reset_index(inplace=True)

def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data['Date'], data['Close'], label='Precio de Cierre', color='blue')

    plt.title(title, fontsize=16)
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Precio (USD)', fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)

    plt.show()

make_graph(tesla_data, "Acciones de Tesla (TSLA) - Precio de Cierre (Últimos 5 Años)")
```
